```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.linear_model import LinearRegression
from sklearn.kernel_ridge import KernelRidge
```

```python
# 非線形の分布になるようにデータを生成

n = 100; N = 2000

x = np.linspace(-3, 3, n)
X = np.linspace(-3, 3, N)

pix = np.pi * x
y = np.sin(pix) / pix + 0.05 * x + 0.1 * np.random.randn(n)

x = x.reshape(-1, 1)
X = X.reshape(-1, 1)
y = y.reshape(-1, 1)

plt.scatter(x,y)
```

```python
# =====================
# 線形回帰でやってみる
# =====================
```

```python
# モデル生成
model = LinearRegression()
model.get_params()
```

```python
model.fit(x, y)
```

```python
p = model.predict(X)

plt.scatter(x, y)
plt.plot(X,p)
```

```python
print(model.score(x, y))
```

```python
# =====================
# 非線形回帰でやってみる
# =====================
```

```python
model2 = KernelRidge(alpha=0.001, kernel='rbf')
```

```python
model2.fit(x, y)
```

```python
p2 = model2.predict(X)

plt.scatter(x, y)
plt.plot(X,p2)
```

```python
print(model2.score(x, y))
```

【考察】
線形回帰、非線形回帰を実際にplotしつつ検証した。
（当然ながら）非線形な分布のデータは線形モデルでは十分なスコアが出ず、非線形モデルであれば十分すぎるスコアを出すことができた。（非線形のほうは若干過学習気味？のように見えるので、パラメータ調整など検証していきたい。
